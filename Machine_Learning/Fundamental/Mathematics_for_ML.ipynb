```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, expon, uniform
import sympy as sp
from ipywidgets import interact

# --- 1. Linear Algebra ---

# 1.1 Vectors
def vector_operations():
    v1 = np.array([2, 3])
    v2 = np.array([1, 4])

    addition = v1 + v2
    dot_product = np.dot(v1, v2)
    cross_product = np.cross([1, 0, 0], [0, 1, 0])
    norm = np.linalg.norm(v1)
    projection = np.dot(v1, v2) / np.linalg.norm(v2)**2 * v2

    print(f"Addition: {addition}")
    print(f"Dot Product: {dot_product}")
    print(f"Cross Product: {cross_product}")
    print(f"Norm of v1: {norm}")
    print(f"Projection of v1 onto v2: {projection}")

# 1.2 Matrices
def matrix_operations():
    A = np.array([[1, 2], [3, 4]])
    B = np.array([[2, 0], [1, 3]])

    multiplication = np.dot(A, B)
    rank = np.linalg.matrix_rank(A)
    trace = np.trace(A)
    det = np.linalg.det(A)
    svd = np.linalg.svd(A)

    print(f"Multiplication:\n{multiplication}")
    print(f"Rank of A: {rank}")
    print(f"Trace of A: {trace}")
    print(f"Determinant of A: {det}")
    print(f"SVD of A:\n{svd}")

# 1.3 Eigenvalues and Eigenvectors
def eigen_operations():
    A = np.array([[1, 2], [3, 4]])
    eigenvalues, eigenvectors = np.linalg.eig(A)
    print(f"Eigenvalues: {eigenvalues}")
    print(f"Eigenvectors:\n{eigenvectors}")

# --- 2. Probability and Statistics ---

# 2.1 Bayes' Theorem
def bayes_theorem(p_b_given_a, p_a, p_b):
    return (p_b_given_a * p_a) / p_b

# 2.2 Probability Distributions
def plot_distributions():
    # Normal Distribution
    x = np.linspace(-3, 3, 100)
    pdf = norm.pdf(x)
    plt.plot(x, pdf, label='Normal Distribution')

    # Exponential Distribution
    pdf_expo = expon.pdf(x)
    plt.plot(x, pdf_expo, label='Exponential Distribution')

    # Uniform Distribution
    pdf_uni = uniform.pdf(x, loc=-2, scale=4)
    plt.plot(x, pdf_uni, label='Uniform Distribution')

    plt.legend()
    plt.title("Probability Distributions")
    plt.show()

# --- 3. Calculus ---

# 3.1 Derivatives
def calculate_derivative():
    x = sp.Symbol('x')
    f = x**3 - 4*x**2 + x + 6
    f_prime = sp.diff(f, x)
    print(f"Derivative of f: {f_prime}")

# --- 4. Optimization ---

# 4.1 Gradient Descent
def gradient_descent(gradient, start, learn_rate, n_iter):
    x = start
    for _ in range(n_iter):
        x -= learn_rate * gradient(x)
    return x

def gradient_descent_examples():
    # Quadratic example
    gradient = lambda x: 2 * x  # Derivative of x^2
    result = gradient_descent(gradient, start=10, learn_rate=0.1, n_iter=100)
    print(f"Minimum value of x: {result}")

    # Logistic regression cost function example
    gradient_log = lambda x: (1 / (1 + np.exp(-x))) - 0.5
    result_log = gradient_descent(gradient_log, start=2, learn_rate=0.1, n_iter=100)
    print(f"Minimum value of logistic regression cost: {result_log}")



# --- Run Examples ---
if __name__ == "__main__":
    print("Linear Algebra: Vectors")
    vector_operations()

    print("\nLinear Algebra: Matrices")
    matrix_operations()

    print("\nLinear Algebra: Eigenvalues and Eigenvectors")
    eigen_operations()

    print("\nProbability Distributions")
    plot_distributions()

    print("\nCalculus: Derivatives")
    calculate_derivative()

    print("\nOptimization: Gradient Descent")
    gradient_descent_examples()

    
```

## Mathematics for Machine Learning

This Python script demonstrates fundamental mathematical concepts used in machine learning, focusing on practical implementations of Linear Algebra, Probability and Statistics, Calculus, and Optimization.

---

## 1. Linear Algebra

### 1.1 Vectors

The `vector_operations()` function demonstrates basic operations on vectors. It creates two vectors, `v1` and `v2`, and computes:

1. **Addition**: The sum of the two vectors.
2. **Dot Product**: A scalar value representing the product of the magnitudes of the vectors and the cosine of the angle between them.
3. **Cross Product**: In 3D, this produces a vector orthogonal to both input vectors. Here, `[1, 0, 0]` and `[0, 1, 0]` result in `[0, 0, 1]`.
4. **Norm**: The magnitude of a vector, computed as the square root of the sum of the squares of its components.
5. **Projection**: The projection of `v1` onto `v2`, calculated using the dot product and the norm of `v2`.

---

### 1.2 Matrices

The `matrix_operations()` function explores operations on matrices `A` and `B`:

1. **Multiplication**: Computes the matrix product using `np.dot`.
2. **Rank**: Determines the number of linearly independent rows or columns.
3. **Trace**: The sum of the diagonal elements.
4. **Determinant**: A scalar value that encodes certain properties of the matrix, including invertibility.
5. **SVD (Singular Value Decomposition)**: Decomposes `A` into three matrices that reveal its structure.

---

### 1.3 Eigenvalues and Eigenvectors

The `eigen_operations()` function calculates the eigenvalues and eigenvectors of matrix `A`:

- **Eigenvalues**: Scalars indicating how much eigenvectors are scaled during the transformation.
- **Eigenvectors**: Non-zero vectors whose direction remains unchanged under the transformation defined by the matrix.

---

## 2. Probability and Statistics

### 2.1 Bayes' Theorem

The `bayes_theorem()` function implements Bayes theorem to calculate the posterior probability `P(A|B)` given:
- `P(B|A)`: Probability of `B` given `A`.
- `P(A)`: Prior probability of `A`.
- `P(B)`: Probability of `B`.

---

### 2.2 Probability Distributions

The `plot_distributions()` function visualizes three key probability distributions:

1. **Normal Distribution**: Symmetrical "bell curve" commonly used in statistics.
2. **Exponential Distribution**: Models time between events in a Poisson process.
3. **Uniform Distribution**: Models equal probability for all outcomes in a range.

Each distribution is plotted on the same graph for comparison.

---

## 3. Calculus

### 3.1 Derivatives

The `calculate_derivative()` function uses the `sympy` library to compute the derivative of a polynomial function \( f(x) = x^3 - 4x^2 + x + 6 \). Derivatives represent the rate of change of a function and are crucial for optimization problems in machine learning.

---

## 4. Optimization

### 4.1 Gradient Descent

The `gradient_descent()` function implements the gradient descent algorithm to minimize a given function. It iteratively updates the variable `x` by subtracting the gradient scaled by a learning rate.

The `gradient_descent_examples()` function demonstrates:

1. **Quadratic Function**: Minimize \( f(x) = x^2 \), whose gradient is \( 2x \).
2. **Logistic Regression Cost Function**: Demonstrates optimization for a sigmoid-based function.

Gradient descent is fundamental in training machine learning models.
